```python
import numpy as np
import pandas as pd
```

```python
cellphones = pd.read_csv('cell_phones.csv')
population = pd.read_csv('population.csv')
country = pd.read_csv('ddf--entities--geo--country.csv')
```

```python
phone_population = pd.merge(cellphones, population, how="inner")
phone_population
```

```python
new_country = country.rename(columns={"country": "geo", "name": "Country"})
```

```python
countryGeo = pd.DataFrame(new_country["geo"])
countryName = pd.DataFrame(new_country["Country"])
name_geo = countryName.join(countryGeo).reset_index(drop=True)
# pd.concat([new_country.geo, new_country.Country])
```

```python
merged = pd.merge(phone_population, name_geo, how="inner")
```

```python
cells_persons = merged["cell_phones_total"] / merged["population"]
cells_persons = pd.DataFrame(cells_persons)
```

```python
country_population_cells = merged.join(cells_persons, how="inner")
```
